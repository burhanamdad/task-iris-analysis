```python
# IRIS DATASET ANALYSIS & VISUALIZATION PROJECT

# IMPORT LIBRARIES

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# PROFESSIONAL PLOT STYLE

sns.set_style("whitegrid")

plt.rcParams['figure.figsize'] = (8,6)

plt.rcParams['font.size'] = 12


# LOAD IRIS DATASET

df = sns.load_dataset('iris')


# BASIC DATASET INFORMATION

print("IRIS DATASET INFORMATION")

print("\nDataset Shape:")
print(df.shape)

print("\nDataset Columns:")
print(df.columns)

print("\nFirst 5 Rows:")
print(df.head())


# DATASET INFO

print("\n===================================")
print("DATASET INFO")
print("===================================")

print(df.info())
print(df.describe())


print("\n===================================")
print("MISSING VALUES")
print("===================================")

print(df.isnull().sum())

# DATA VISUALIZATION
# SCATTER PLOT

plt.figure(figsize=(8,6))

sns.scatterplot(
    data=df,
    x='sepal_length',
    y='sepal_width',
    hue='species',
    s=100
)

plt.title(
    "Sepal Length vs Sepal Width",
    fontsize=16,
    fontweight='bold'
)

plt.xlabel("Sepal Length")

plt.ylabel("Sepal Width")

plt.savefig(
    "scatter_plot.png",
    bbox_inches='tight'
)

plt.show()

# HISTOGRAMS

df.hist(
    figsize=(12,10),
    bins=15
)

plt.suptitle(
    "Feature Distributions",
    fontsize=18,
    fontweight='bold'
)

plt.savefig(
    "histograms.png",
    bbox_inches='tight'
)

plt.show()


# BOX PLOT

plt.figure(figsize=(10,6))

sns.boxplot(
    data=df,
    palette='Set2'
)

plt.title(
    "Boxplot of Iris Features",
    fontsize=16,
    fontweight='bold'
)

plt.savefig(
    "boxplot.png",
    bbox_inches='tight'
)

plt.show()


# PAIRPLOT


sns.pairplot(
    df,
    hue='species'
)

plt.savefig(
    "pairplot.png",
    bbox_inches='tight'
)

plt.show()


print("\nGenerated Files:")

print("1. scatter_plot.png")

print("2. histograms.png")

print("3. boxplot.png")

print("4. pairplot.png")
```
